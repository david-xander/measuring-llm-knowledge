```python
# EXCLUDE keywords
excluded_keywords = ['!', '"', '#', '%', '&', '(', ')', '*', '+', ',', '-', '--', '.', '..', '...', '/', '//', ':', '::', ';', '<', '<<', '<=', '=', '==', '>', '>=', '>>', '[', "'", '\\', '\n', '\r', ']', '^', 'u{', '{', '|', '}', '~', '~=']
   
# GROUP keywords
grouped_keywords = {'Control Flow': ['break', 'continue', 'do', 'else', 'elseif', 'end', 'for', 'if', 'repeat', 'return', 'then', 'while'], 'Logical Operators': ['and', 'not', 'or'], 'Data Types': ['false', 'nil', 'true'], 'Declarations': ['function', 'local', 'goto', 'x'], 'Loop Constructs': ['repeat', 'for', 'while'], 'Special Characters': ['0', "'", '\\', '\n', '\r']}
df = ResultsDataset().filter_and_group_keywords(df, excluded_keywords, None)
df.head()                    
        
```
