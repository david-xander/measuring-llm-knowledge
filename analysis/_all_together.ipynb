```python
 
import sys
sys.path.append('../')                            
from core.results_analysis import ResultsDataset     
from core.results_analysis import ResultsGraph
from core.results_analysis import ResultsCoverage
from core.generate_keyword_features import KeywordGroupingGenerator 

from core.analysis_lib import RunAnalysis

experimented_languages = [
    {
        "language": "ocl", 
        "datasets": ["BaseLine","deepseek","deepseek-sm","gpt-4o","gpt-4o-mini","llama","llama-sm"], 
        "description": "Abukhalaf et al., 2023",
        "result": None
    }, 
    {
        "language": "plantuml",
        "datasets":["BaseLine","deepseek","deepseek-sm","gpt-4o","gpt-4o-mini","llama","llama-sm"],
        "description": "GoldenUMLmodelset",
        "result": None
    },
    {
        "language": "r",
        "datasets":['deepseek-he', 'deepseek-sm-he','gpt-4o-he', 'gpt-4o-mini-he', 'llama-he', 'llama-sm-he'],
        "description": "Multipl-E(HumanEval)",
        "result": None
    }, 
    {
        "language": "r",
        "datasets":['deepseek-mbpp', 'deepseek-sm-mbpp', 'gpt-4o-mbpp', 'gpt-4o-mini-mbpp', 'llama-mbpp', 'llama-sm-mbpp'],
        "description": "Multipl-E(MBPP)",
        "result": None
    },        
    {
        "language": "lua",
        "datasets":['deepseek-he', 'deepseek-sm-he', 'gpt-4o-he', 'gpt-4o-mini-he', 'llama-he', 'llama-sm-he'],
        "description": "Multipl-E(HumanEval)",
        "result": None
    }, 
    {
        "language": "lua",
        "datasets":['deepseek-mbpp', 'deepseek-sm-mbpp', 'gpt-4o-mbpp', 'gpt-4o-mini-mbpp', 'llama-mbpp', 'llama-sm-mbpp'],
        "description": "Multipl-E(MBPP)",
        "result": None
    },
    {
        "language": "sql",
        "datasets":['BaseLine','deepseek', 'deepseek-sm', 'gpt-4o', 'gpt-4o-mini', 'llama', 'llama-sm'],
        "description": "Spider1.0",
        "result": None
    },   
    {
        "language": "python3",
        "datasets":['BaseLine-HumanEval', 'deepseek-he', 'deepseek-sm-he', 'gpt-4o-mini-he', 'gpt-4o-he', 'llama-he', 'llama-sm-he'],
        "description": "HumanEval",
        "result": None
    },
    {
        "language": "python3",
        "datasets":['BaseLine-MBPP', 'deepseek-mbpp', 'deepseek-sm-mbpp','gpt-4o-mini-mbpp', 'gpt-4o-mbpp', 'llama-mbpp', 'llama-sm-mbpp'],
        "description": "MBPP",
        "result": None
    },            
]


for item in experimented_languages:
    language = item["language"]
    datasets = item["datasets"]

    kwgenerator = KeywordGroupingGenerator(
        language=language,
        keywords=None,
    )

    excluded_keywords = kwgenerator.get_excluded_keywords()
    grouped_keywords = kwgenerator.get_semantic_groups()


    df = ResultsDataset().load_dataset_for_language(language)

    grouped_keywords = ResultsDataset().check_for_group_consistency(df, grouped_keywords, excluded_keywords)

    # Load datasets
    df = ResultsDataset().load_dataset_for_language(language)
    syntactically_correct_kw_ds, total_kw_ds, rule_usage_vectors_ds, rule_correct_vectors_ds, rule_usage_accuracy_ds, rule_errors_ds, parser_lexer_errors, tree_depth_and_size, sequences_ds = ResultsDataset().load_syntactic_analysis_dataset_for_language(language)

    # Exclude columns not relevant for analysis, but NO GROUPING
    df = ResultsDataset().filter_and_group_keywords(df, excluded_keywords, None)
    syntactically_correct_kw_ds = ResultsDataset().filter_and_group_keywords(syntactically_correct_kw_ds, excluded_keywords, None)
    total_kw_ds = ResultsDataset().filter_and_group_keywords(total_kw_ds, excluded_keywords, None)


    grouped_tables = ResultsCoverage(df).get_grouped_tables(df, syntactically_correct_kw_ds, total_kw_ds, grouped_keywords)



    res = ResultsCoverage(df).get_keyword_coverage_and_correctness_table_v2(df, syntactically_correct_kw_ds, total_kw_ds, rule_usage_vectors_ds, rule_correct_vectors_ds, tree_depth_and_size, datasets)

    item["result"] = res.to_dict()
```
